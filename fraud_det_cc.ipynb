```python
fil=ps.read_csv('creditcard.csv')
```

```python
fil=fil.drop(['Amount','Time'],axis=1)
```

```python
fil.head()
```

```python
df_0=fil[fil['Class']==0]
df_1=fil[fil['Class']==1]
```

```python
df_0.shape
```

```python
dat=ps.concat([df_0[:508],df_1],axis=0)
```

```python
dat.shape
```

```python
x=dat.drop('Class',axis=1)
y=dat['Class']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
lg.fit(x_train,y_train)
```

```python
pre=lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,roc_auc_score
```

```python
acc_lr = accuracy_score(y_test,pre)
conf = confusion_matrix(y_test,pre)
clf_report = classification_report(y_test,pre)
print(f"Accuracy Score of Logistic Regression : {acc_lr}")
print(f"Confusion Matrix : \n{conf}")
print(f"Classification Report : \n{clf_report}")
```
